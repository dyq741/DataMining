```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
# 

# 

data2_dropped = data2.dropna()
print(data2_dropped)
print('Dropped Rows:', data2.shape[0] - data2_dropped.shape[0])

# 

data2_filled = data2.fillna(data2.mode().iloc[0])
print(data2_filled)
print('Filled Rows:', data2_filled.isna().sum())
```

```python
#
data_fill2 = data2
corr_matrix = data_fill2.corr()
most_correlated = corr_matrix["Data_Value"].drop("Data_Value").idxmax()
mean_value = data_fill2[most_correlated].mean()
data_fill2["Data_Value"].fillna(mean_value,inplace=True)
print('Filled Rows:', data_fill2.isna().sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.impute import KNNImputer
from sklearn.linear_model import LinearRegression
%matplotlib inline
x = data2[["YearStart","YearEnd","LocationID"]]
y = data2['Data_Value']
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2, test_size=0.25)
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.fit_transform(x_test)
param_grid = {"n_neighbors": [3, 5, 7, 9]}
estimator = KNeighborsClassifier()
estimator = GridSearchCV(estimator=estimator, param_grid=param_grid, cv=10, n_jobs=4)
estimator.fit(x_train, y_train.astype('string'))
score_ret = estimator.score(x_test, y_test.astype('string'))
y_pre = estimator.predict(x_test)
print(":\n", estimator.best_estimator_)
print(":\n", estimator.best_score_)
print(":\n", estimator.cv_results_)
```
